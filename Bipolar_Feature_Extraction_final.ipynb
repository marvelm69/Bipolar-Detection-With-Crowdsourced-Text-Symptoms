Tokenization + Lemmatizing

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
nltk.download('averaged_perceptron_tagger')
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
#before lem
dff[['id', 'q_11']].head()
```

```python
import spacy

# Memuat model bahasa Inggris
nlp = spacy.load("en_core_web_sm")

# Fungsi untuk melakukan lemmatisasi pada teks
def lemmatize_text1(text):
    # Membuat dokumen spaCy dari teks
    doc = nlp(text)
    # Menggabungkan kata-kata yang telah dilemmatisasi
    lemmatized_text = [token.lemma_ for token in doc]
    return ' '.join(lemmatized_text)
# Load the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Function to map NLTK POS tags to WordNet POS tags
def pos_tagger(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:
        return None

# Function to lemmatize the text
def lemmatize_text2(text):
    # Tokenize the text
    tokens = nltk.word_tokenize(text)

    # Get the POS tags for the tokens
    nltk_tags = nltk.pos_tag(tokens)

    # Map NLTK POS tags to WordNet POS tags and lemmatize
    lem_text = []
    for word, tag in nltk_tags:
        wn_tag = pos_tagger(tag)
        if wn_tag is None:
            lem_text.append(word)
        else:
            lem_text.append(lemmatizer.lemmatize(word, wn_tag))

    # Join the lemmatized tokens into a string
    lem_text_str = ' '.join(lem_text)

    return lem_text_str
```

```python
for i in range(10):
  dff['q_11'] = dff['q_11'].apply(lemmatize_text1)
  dff['q_11'] = dff['q_11'].apply(lemmatize_text2)
```

Remove Stop Words

```python
nltk.download('stopwords')
```

```python
# Mengambil daftar stopwords dalam bahasa Inggris
stopwords = nltk.corpus.stopwords.words('english')

# Fungsi untuk menghapus stopwords dari teks
def remove_stopwords(text):
    # Tokenisasi teks
    word_tokens = nltk.word_tokenize(text)
    # Menghapus stopwords dari teks
    filtered_tokens = [word for word in word_tokens if word.lower() not in stopwords]
    # Menggabungkan kembali kata-kata yang tersisa
    filtered_text = ' '.join(filtered_tokens)
    return filtered_text

# Menghapus stopwords dari kolom 'q_11' dalam dataframe 'dff'
for i in range(len(dff['q_11'])):
    dff['q_11'][i] = remove_stopwords(dff['q_11'][i])

# Menampilkan dataframe 'dff' setelah stopwords dihapus dari kolom 'q_11'
dff[['id', 'q_11']].head(10)
```

```python
# Membaca list kata-kata dari file CSV
valid_words = set()
with open("list_temp_lemmed (1).csv", "r") as file:
    for line in file:
        synonyms = line.strip().split("  ")  # Pisahkan sinonim dengan double space
        valid_words.update(synonyms)

# Fungsi untuk membuang kata-kata yang tidak ada dalam valid_words
def remove_invalid_words(text):
    # Tokenisasi teks
    tokens = text.split()
    # Pilih kata-kata yang terdapat dalam valid_words
    valid_tokens = [token for token in tokens if token in valid_words]
    # Gabungkan kembali kata-kata yang valid menjadi sebuah string
    valid_text = ' '.join(valid_tokens)
    return valid_text

# Terapkan fungsi remove_invalid_words ke kolom 'q_11' dalam dataframe 'dff'
dff['q_11'] = dff['q_11'].apply(remove_invalid_words)

# Tampilkan dataframe 'dff' setelah menghapus kata-kata yang tidak valid
dff[['id', 'q_11']].head(10)
```

```python
def tihuan_tongyici(string1):
    combine_dict = {}
    with open("list_temp_lemmed.txt", "r") as file:
        for line in file:
            separate_word = line.strip().split(",")
            for word in separate_word[1:]:
                combine_dict[word] = separate_word[0]

    final_sentence = " ".join(combine_dict.get(word, word) for word in string1.split())
    return final_sentence

# Menggunakan fungsi tihuan_tongyici untuk mengubah kolom 'q_11' dari dataframe 'dff'
dff['q_11'] = dff['q_11'].apply(tihuan_tongyici)

# Menampilkan beberapa baris pertama dari dataframe dff setelah pengubahan
dff[['id', 'q_11']].head(10)
```

Vectorizing

```python
from sklearn.feature_extraction.text import CountVectorizer

# Membuat objek CountVectorizer dengan binary=True
binary_vectorizer = CountVectorizer(binary=True)

# Melakukan vektorisasi pada kolom 'q_11' dalam dataframe 'dff'
X_binary = binary_vectorizer.fit_transform(dff['symptoms_extracted'])

# Mendapatkan daftar fitur yang dihasilkan
feature_names_binary = binary_vectorizer.get_feature_names_out()

# Membuat dataframe dari matriks vektorisasi
symptoms = pd.DataFrame(X_binary.toarray(), columns=feature_names_binary)

# Menampilkan dataframe fitur biner
symptoms.head()
```
